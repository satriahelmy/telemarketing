# 2. Load Train Data

```python
x_train = joblib.load("data/processed/x_train.pkl")
y_train = joblib.load("data/processed/y_train.pkl")
```

```python
#gabungkan x dan y train
dataset = pd.concat([x_train, y_train], axis = 1)
```

# 3. EDA

## 3.1 Cek Missing Values

```python
dataset.isnull().sum()
```

## 3.2. Cek Informasi Statistik

### 3.2.1. Cek Skewness

```python
dataset.skew(numeric_only = True)
```

### 3.2.2. Pemisahan Dataset berdasarkan Kelas

```python
# pisahkan menjadi 2 kelas untuk melihat apakah kita perlu melakukan imputasi berdasarkan kelas
dataset_yes = dataset[dataset['y'] == "yes"].copy()
dataset_no = dataset[dataset['y'] != "yes"].copy()
```

```python
dataset_yes.describe()
```

```python
dataset_no.describe()
```

### 3.2.3. Visualisasi Fitur Age

```python
sns.kdeplot(data = dataset, x = "age")
plt.axvline(dataset['age'].mean(), 0, 1, color = "g")
plt.axvline(dataset['age'].median(), 0, 1, color = "r")
plt.show()
```

```python
sns.kdeplot(data = dataset, x = "age", hue = "y")
plt.axvline(dataset_yes['age'].mean(), 0, 1, color = "g")
plt.axvline(dataset_no['age'].mean(), 0, 1, color = "k")
plt.show()
```

### 3.2.3. Visualisasi Fitur Duration

```python
sns.kdeplot(data = dataset, x = "duration")
plt.axvline(dataset['duration'].mean(), 0, 1, color = "g")
plt.axvline(dataset['duration'].median(), 0, 1, color = "r")
plt.show()
```
